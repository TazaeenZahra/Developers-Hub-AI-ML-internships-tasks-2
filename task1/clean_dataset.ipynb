```python
df.drop(columns='customerID',inplace=True)
```

```python
df.shape
```

```python
df['MultipleLines'] = df['MultipleLines'].str.replace(
    'No phone service',
    'No'
)
```

```python
df[df['TotalCharges'].str.isnumeric()]['TotalCharges'].astype(float)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['TotalCharges'] = df['TotalCharges'].fillna(0)

print(df['TotalCharges'].dtype)
```

```python
df['OnlineSecurity'] = df['OnlineSecurity'].str.replace(
    'No internet service',
    'No'
)
```

```python
df['OnlineBackup'] = df['OnlineBackup'].str.replace(
    'No internet service',
    'No'
)
```

```python

df['DeviceProtection'] = df['DeviceProtection'].str.replace(
    'No internet service',
    'No'
)
```

```python

df['TechSupport'] = df['TechSupport'].str.replace(
    'No internet service',
    'No'
)
```

```python

df['StreamingTV'] = df['StreamingTV'].str.replace(
    'No internet service',
    'No'
)
```

```python

df['StreamingMovies'] = df['StreamingMovies'].str.replace(
    'No internet service',
    'No'
)
```
